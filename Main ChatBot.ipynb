```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.lancaster import LancasterStemmer
from sklearn import preprocessing
import nltk
import json
```

```python
#Reading in the file using json so I can be able to access it as a dict
import json
file = r'C:\Users\Stina\Documents\LEARNING PYTHON\Data I work with\ChatBot-master\chat.json'
file= file.lstrip('\u202a')
with open(file) as f:
    data=json.load(f)
data
```

```python
words=[]
tag=[]
```

```python
for  intent in data['intents']:
    for pattern in intent['patterns']:
        #words = nltk.word_tokenize(pattern) #this code won't work cox I need internet connection
        words.append(pattern)
        tag.append(intent['tag'])
```

```python
#stemming the word to the root word
Stemmer = LancasterStemmer()
training = [Stemmer.stem(i) for i in words]
```

```python
training
```

```python
tag
```

```python
#Here, we learn the vocabulary of the word and turn it into an array
vect = CountVectorizer()
x_train=vect.fit_transform(training)
```

```python
x_train
```

```python
#Converting the categorical data to numerical data using one hot encoding
df_tag=pd.get_dummies(tag)
```

```python
y_train=np.array(df_tag)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Dense(128,input_shape=(83,),activation ='relu'),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(64,activation ='relu'),
                          keras.layers.Dense(48,activation = 'relu'),
                          
                          keras.layers.Dense((14),activation='softmax')])

model.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=200, batch_size = 8, verbose =1)
```

```python
#Saving model
model.save('my_model')
```

```python
model = keras.models.load_model("my_model")  #Loading the model after saving it
```

```python
labels = np.hstack(df_tag)
```
